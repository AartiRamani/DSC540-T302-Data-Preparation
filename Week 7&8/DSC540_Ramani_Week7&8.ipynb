## Chapter 8

### Combine and Merge Datasets (you will have to either create a new dataset from your existing data or create a relationship between the data I have provided)

```python
#Merging the 2015 & 2016 candy dataframes 
merged_df = pd.merge(candy_2015_df,candy_2016_df,on="How old are you?",how="left",suffixes=('_2015','_2016'))
print(merged_df.shape)
merged_df.head(5)
```

### Pivot the data

```python
#Using pivot function on a column in the merged dataframe
merged_df.pivot_table(values=[' [Butterfinger]_2015',' [Butterfinger]_2016'], index='How old are you?', dropna=True, aggfunc='first')
```

## Chapter 11

### Convert between string and date time

```python
#data type of Timestamp_2015 column in the merged dataframe
merged_df = merged_df.reset_index()
merged_df['Timestamp_2015']
```

```python
#Converting the DateTime column to string
merged_df['Timestamp'] = merged_df['Timestamp_2015'].dt.strftime('%Y-%m-%d %H:%M') 
```

```python
#Validate the datatype after converting to string
merged_df['Timestamp']
```

```python
merged_df.head(5)
```

### Convert timestamps to periods and back

```python
#Converting the timestamp to period (by quarter) and assigning to a new column in the dataframe
merged_df["Period"]=pd.DatetimeIndex(merged_df["Timestamp"]).to_period('Q')
```

```python
merged_df.head(5)
```

```python
#To convert back to timestamps, use to_timestamp
merged_df['Period'] = merged_df.Period.dt.to_timestamp(how='end')
 
merged_df.head(5)
```
